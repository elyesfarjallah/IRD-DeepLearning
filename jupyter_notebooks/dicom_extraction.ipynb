```python
import os

def find_files(folder_path):
    """
    Recursively find all files in a folder and its subfolders.
    
    Args:
    - folder_path (str): Path to the folder to search.
    
    Returns:
    - file_paths (list): List of paths to all files found.
    """
    file_paths = []
    for root, dirs, files in os.walk(folder_path):
        for file in files:
            file_paths.append(os.path.join(root, file))
    return file_paths

# Example usage:
folder_path = "/home/efarjall/ird_dataset/export_heyex_original_dataset_03/DICOM/0001"
all_files = find_files(folder_path)
print(all_files)
```
